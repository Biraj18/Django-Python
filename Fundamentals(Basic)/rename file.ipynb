# Making folder and moving data to respective folder 

```python
import shutil
src='E:\mbl phots and songs\python modified'  #source folder
os.chdir(src)                                 #changing directory
modify=os.listdir(src)                        #listing all directory files

valid_ext=['jpg','png']
if not os.path.exists(src+'\Photos'):
    os.mkdir('Photos')
for i,file in enumerate(modify):
    ext=file.split('.')[-1]
    if ext.lower() in valid_ext:
        os.rename(file,f'Hetauda_{i}.{ext}')  #renames function can also be used to rename and move file
        shutil.move(src+ f'\\Hetauda_{i}.{ext}',src+'\Photos') 
```
